<h1>Using Chordify to Annotate Intervals<h1>

One great way to quickly make a reduction of a score is with *chordify* and the *annotateIntervals* method on Chords. We can use one of the multipart-opus ABC files described in the last chapter to demonstrate.

Let us load up one of the most beautiful memorial pieces of all time, the motet on the death of Johannes Ockeghem by Josquin des Prez (d. 1521):

```python
from music21 import * #importing all functions from Music21
o = corpus.parse('josquin/laDeplorationDeLaMorteDeJohannesOckeghem') #this parses the entire piece into the variable
#o as a stream containing all the notes
for s in o: #This is a simple for loop that prints all the scores inside o, in this case 5 scores
    print(s)
```

Okay so we have a collection of scores to merge into parts – this is an unusual but not totally rare phenomenon, so we merge them:

```python
mergedScores = o.mergeScores() #this merges the 5 scores seen above into one, storing it in the variable mergedScores
scoreExcerpt = mergedScores.measures(127, 133) #this takes the measures 127 to 133 and stores them in the variable scoreExcerpt
scoreExcerpt.show() #the chosen measures are shown as notation now
```

Let’s chordify it:

```python
reduction = scoreExcerpt.chordify() #the excerpt above is chordified and stored in the variable reduction
reduction.show() #the reduced/chordified version is shown as notation
```

We’ll iterate over the chords and put them in closed position (we can probably ask Sebastian about this, as non-musicians we were a little confused about what this does) in octave 4 and run the *annotateIntervals* command:

```python
for c in reduction.recurse().getElementsByClass('Chord'): #this for loop runs through the stream and only
    #affects the objects of the class 'Chord' via the getElementsByClass 
    c.closedPosition(forceOctave=4, inPlace=True) #this puts the notes in closed position (i don't really know what that means, ask Sebastion)
    c.annotateIntervals() #this just annotates the intervals in the way described above
```

We will put the reduction back into the score and show it. We insert it at the zero point of the score, rather than using append because it begins at the same time point as the other parts. Let’s also get rid of the fourth part, since it’s blank.

```python
scoreExcerpt.insert(0, reduction) #this inserts the reduction we created above into the stream in scoreExcerpt at the zero point
emptyPart = scoreExcerpt.parts[3] #this puts the 4th part of the score, which only consisted of pauses in the given measures into the variable emptyPart
scoreExcerpt.remove(emptyPart) #this removes the empty part we declared in the line above
scoreExcerpt.show() #the excerpt is shown again with all the modifications we did 
```

The intervals have been added as lyrics on the chord:

```python
for c in reduction.recurse().getElementsByClass('Chord'): #this for loop runs through the reduction stream and identifies the Chord Elements
    print(c, end=" ") 
    for l in c.lyrics:
        print(l.text, end=" ")
    print()
```

From here it is easy to find interesting places like that “4321” chord at the end of the fifth measure of the excerpt, somthing I didn’t know would be found in high Renaissance polyphony even as a collection of passing tones! This is something that probably makes a lot more sense for people who actually know shit about music.

<h2>Chordify and advanced scores</h2>

More complex scores can also be chordified. If there are lots of tuplets, you might get odd results. Such as with Opus 19, no. 6, by Arnold Schoenberg.

```python
schoenberg = corpus.parse('schoenberg/opus19', 6) #this parses another piece of music (its sixth measure to be exact) into the variable schoenberg
schoenberg.show() #the stream is displayed as notation
```

If you thought this was chaotic, then watch this:

```python
chorded = schoenberg.chordify() #this chordifies the complex piece of music seen above
chorded.show() #the cordified version is shown
```

There are more specialized commands for *.chordify*, so if you want to learn more, look at the __chordify()__ documentation. We will get to the option, *addPartIdAsGroup* later, which will let you know exactly where each pitch in the chordified Chord comes from.
